```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('water_potability.csv')
df.head()
```

```python
df2 = df.copy()
df2.fillna(0,inplace = True)
mean = []
cols = list(df.columns)
for x in cols:
    mean.append(df2[x].mean())
```

```python
print(df2.shape)
df2.head()
```

```python
i = 0
for x in cols:
    df[x].fillna(mean[i],inplace= True)
    i += 1
```

```python
df.describe()
```

```python
from tensorflow.keras import layers, models
from tensorflow.keras import regularizers
from tensorflow import keras
import numpy as np
```

```python
model = models.Sequential()
model.add(layers.Dense(64, activation='relu', input_shape=(len(cols)-1,)))
model.add(layers.Dropout(0.25))
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dropout(0.25))
model.add(layers.Dense(1, activation='sigmoid'))
```

```python
opt_names= { 'sgd':'sgd',
            'adam' : keras.optimizers.Adam(learning_rate=0.0003),
            'rmsprop': keras.optimizers.RMSprop(learning_rate=0.1),
            'adadelta': keras.optimizers.Adadelta(learning_rate=0.001, rho=0.95, epsilon=1e-07, name="Adadelta"),
            'adagrad' : keras.optimizers.Adagrad(learning_rate=0.001, initial_accumulator_value=0.1, epsilon=1e-07, name="Adagrad"),
        'adamax' : keras.optimizers.Adamax(learning_rate=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-07, name="Adamax",), 
    'nadam':keras.optimizers.Nadam(learning_rate=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-07, name="Nadam")}


```

```python
from sklearn.model_selection import train_test_split
X = df.drop('Potability', axis=1)
y = df['Potability']
dt_features = StandardScaler().fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(dt_features, y, test_size=0.20, random_state=1)
```

```python
'''
mean = X_train.mean(axis=0)
X_train = X_train - mean
std = X_train.std()
X_train /= std
y_train -= mean
y_train /= std
'''
```

```python
data, x_val, target, y_val = train_test_split(X_train, y_train , test_size=0.20, random_state=1)
```

```python
hist = []
```

```python
for name in opt_names:
    opt = opt_names[name]
    print(name)
    model.compile(optimizer=opt,
              loss='binary_crossentropy', #binary_crossentropy
              metrics=['acc']   
             )
    history = model.fit(data,
                    target,
                    epochs=50,
                    batch_size=124,
                    validation_data=(x_val,y_val), verbose = 0)
    hist.append([name, history])
```

```python

for h in hist:
    print(h[0])
    print('loss: %f Acc: %f ----- val_loss: %f, val_acc: %f '
          %((h[1]).history['loss'][-1], (h[1]).history['acc'][-1],(h[1]).history['val_loss'][-1], (h[1]).history['val_acc'][-1]))
```

```python
model.compile(optimizer= keras.optimizers.Adam(learning_rate=0.0003),
              loss='binary_crossentropy', #binary_crossentropy
              metrics=['acc']   
             )
history = model.fit(data,
                    target,
                    epochs=100,
                    batch_size=124,
                    validation_data=(x_val,y_val))
```

```python
history_dict = history.history
loss_values = history_dict['loss']
val_loss_values = history_dict['val_loss']


fig = plt.figure(figsize=(10,10))
epoch = range(1,len(loss_values)+1)
plt.plot(epoch,loss_values, '--',label='loss')
plt.plot(epoch,val_loss_values, '--',label='val_loss')
plt.legend()
plt.show()
```
